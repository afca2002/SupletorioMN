```python
import numpy as np
import matplotlib.pyplot as plt


A = np.array([[1, 1], [-2, 5]], dtype=float)
b = np.array([7, 0], dtype=float).reshape(-1, 1)
x0 = np.array([1, 1], dtype=float).reshape(-1, 1)
tol = 1e-6
max_iter = 100

# Mtodo de Gauss-Jacobi
def gauss_jacobi(A, b, x0, tol, max_iter):
    n = A.shape[0]
    x = x0.copy()
    trajectory = [x.flatten()]
    
    for k in range(max_iter):
        x_new = np.zeros_like(x)
        for i in range(n):
            suma = sum(A[i, j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - suma) / A[i, i]
        
        trajectory.append(x_new.flatten())
        
        if np.linalg.norm(x_new - x) < tol:
            return x_new, trajectory
        
        x = x_new.copy()
    
    return x, trajectory

# Mtodo de Gauss-Seidel
def gauss_seidel(A, b, x0, tol, max_iter):
    n = A.shape[0]
    x = x0.copy()
    trajectory = [x.flatten()]
    
    for k in range(max_iter):
        x_new = x.copy()
        for i in range(n):
            suma = sum(A[i, j] * x_new[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - suma) / A[i, i]
        
        trajectory.append(x_new.flatten())
        
        if np.linalg.norm(x_new - x) < tol:
            return x_new, trajectory
        
        x = x_new.copy()
    
    return x, trajectory

x_jacobi, traj_jacobi = gauss_jacobi(A, b, x0, tol, max_iter)
x_seidel, traj_seidel = gauss_seidel(A, b, x0, tol, max_iter)
traj_jacobi = np.array(traj_jacobi)
traj_seidel = np.array(traj_seidel)
plt.figure(figsize=(10, 6))
plt.plot(traj_jacobi[:, 0], traj_jacobi[:, 1], 'o-', label='Gauss-Jacobi')
plt.plot(traj_seidel[:, 0], traj_seidel[:, 1], 's-', label='Gauss-Seidel')
plt.plot(5, 2, 'rx', label='Solucin exacta')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Trayectoria de Gauss-Jacobi y Gauss-Seidel')
plt.grid(True)
plt.show()
```
